```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Beolvasás és magyar adatok kiválogatása
dataset = pd.read_csv("survey_results_public.csv")
dataset_hungary = dataset.loc[dataset["Country"] == "Hungary"]
del dataset
dataset_hungary["Country"].value_counts(dropna=False)
```

```python
# Üres értékek törlése
dataset_hungary["ConvertedCompYearly"] = dataset_hungary["ConvertedCompYearly"].dropna()
# USD konvertálása HUF-ra (2024 május közepén közel 360 Ft volt egy $)
dataset_hungary["ConvertedCompYearly"] = dataset_hungary["ConvertedCompYearly"] * 360
dataset_hungary.dropna(subset=["ConvertedCompYearly"],inplace=True)
```

```python
dataset_hungary["YearsCodePro"] = dataset_hungary["YearsCodePro"].replace({"Less than 1 year": 0.5, "More than 50 years": 50})
dataset_hungary["YearsCodePro"] = pd.to_numeric(dataset_hungary["YearsCodePro"])
dataset_hungary.dropna(subset=["YearsCodePro"],inplace=True)
```

```python
# Adatok kiszűrése, hogy pontosabb legyen a kimutatás
szurt_adatok = dataset_hungary[
    dataset_hungary["ConvertedCompYearly"].between(1000000,30000000)
]

# Tengelyek létrehozása és a vászon méretezése
fig, ax = plt.subplots(figsize=(6,6))

# Pontdiagram
ax.scatter(szurt_adatok["YearsCodePro"],szurt_adatok["ConvertedCompYearly"],alpha=0.7)

# Tengelyek és a diagram elnevezése
ax.set_title("Éves bruttó fizetés a szakmai tapasztalat (IT) függvényében", fontsize=14, pad=15)
ax.set_xlabel("Szakmai tapasztalat (év)", fontsize=12)
ax.set_ylabel("Éves bruttó fizetés (HUF)", fontsize=12)

# Normálalak kikapcsolása
ax.ticklabel_format(style='plain', axis='y')

plt.show()
```
