```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Beolvasás és magyar adatok kiválogatása
dataset = pd.read_csv("survey_results_public.csv")
dataset_hungary = dataset.loc[dataset["Country"] == "Hungary"]
del dataset
dataset_hungary["Country"].value_counts(dropna=False)
```

```python
# Üres adatok törlése
dataset_hungary.dropna(subset=["RemoteWork"],inplace=True)
# Hybrid érték leegyszerűsítése
dataset_hungary["RemoteWork"] = dataset_hungary["RemoteWork"].replace("Hybrid (some remote, some in-person)","Hybrid")
```

```python
# Tengelyek
fig, ax = plt.subplots(figsize=(6,6))

# Munkavégzés változókká bontása
counts = dataset_hungary["RemoteWork"].value_counts()

munkavegzes = counts.index
darabszam = counts.values

# Kördiagram
ax.pie(darabszam, labels=munkavegzes,autopct='%1.1f%%')
ax.set_title("Munkavégzési formák megoszlása Magyarországon")

plt.show()
```
