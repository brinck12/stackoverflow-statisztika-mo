```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Beolvasás és magyar adatok kiválogatása
dataset = pd.read_csv("survey_results_public.csv")
dataset_hungary = dataset.loc[dataset["Country"] == "Hungary"]
dataset_hungary["Country"].value_counts(dropna=False)
del dataset
```

```python
# Üres értékek törlése
dataset_hungary["ConvertedCompYearly"] = dataset_hungary["ConvertedCompYearly"].dropna()
# USD konvertálása HUF-ra (2024 május közepén közel 360 Ft volt egy $)
dataset_hungary["ConvertedCompYearly"] = dataset_hungary["ConvertedCompYearly"] * 360
```

```python
# Új oszlop létrehozása, az éves bruttó fizetést elosztva 12-vel megkapjuk a havi bruttó fizetéseket
dataset_hungary["HaviFizetes"] = dataset_hungary["ConvertedCompYearly"] / 12
dataset_hungary["HaviFizetes"].value_counts()
```

```python
# Adatok kiszűrése, anélkül pontatlan a statisztika
szurt_adatok = dataset_hungary[
    (dataset_hungary["HaviFizetes"] > 100000) & 
    (dataset_hungary["HaviFizetes"] < 3000000)
]

# Tengelyek
ax = plt.subplot()

# Hisztogram, a bins paraméterrel tudjuk megadni hány sávra ossza fel a vásznat
ax.hist(szurt_adatok["HaviFizetes"], bins=20, edgecolor='black')

# Tengelyek és hisztogram elnevezése
ax.set_xlabel("Havi Bruttó Fizetés (HUF)")
ax.set_ylabel("Gyakoriság (Darabszám)")
ax.set_title("Magyarországi IT fizetések eloszlása")

ax.ticklabel_format(style='plain', axis='x')

plt.tight_layout()
plt.show()
```
